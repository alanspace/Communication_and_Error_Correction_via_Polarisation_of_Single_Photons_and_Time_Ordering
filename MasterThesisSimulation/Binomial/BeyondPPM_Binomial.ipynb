# Combination

$$\binom{n}{r} = \frac{n!}{r!(n-r)!}$$

Suppose we have a list [1,2,4], 3 photons with length of time bins 7.
The ways we can organize them is $3!$, and a selection of these 3 photons from a collection of this 7 time bins in which the order of selection does not matter is $\binom{n}{r}$. \\

https://www.math.net/combination



```python
def combination(n,r):
  return math.factorial(n) / (math.factorial(n-r)*math.factorial(r))

combination(3,2), combination(7,3), combination(14,4)
```

## Binomial Distribution of Loss Error
There are n photons in a time bin. The total probability of lossing $l$ photons and not losing $n-l$ photons

$$P(n, l, P_l) = \binom{n}{l} P_l^l(1-P_l)^{n-l} $$

$n$ is the number of photon. \\
$l$ is the number of loss photon. \\
$P_l$ is the probability of loss error. \\

```python
def LossDistribute(n,l,P_l):
  return combination(n,l)*(P_l)**l*(1-P_l)**(n-l)
```

# Binomial Distribution of Error Probability

$$\text{ Binomail Distribution of Error} = \text{Binomial Distribution of Loss Error} \times \text{Binomial Distribution of Added Error} $$

\\

$$Prob(l, a, n, N, P_l, P_a) = \binom{n}{l} P_l^l(1-P_l)^{n-l} \times \binom{N-n}{a} P_a^a (1-P_a)^{N-n-a}$$

$n$ = number of photon \\
$l$ = number of photon lost \\
$a$ = number of photon added \\
$N$ = length of time bin \\
$P_l$ = probability of lost error \\
$P_a$ = probability of added error \\

```python
def LossDistribute(n,l,P_l):
  return combination(n,l)*(P_l)**l*(1-P_l)**(n-l)  

def AddDistribute(N, n, a, P_a):
  return combination(N-n,a)*(P_a)**a*(1-P_a)**(N-n-a)

def Perr(N,n,l,a, P_l, P_a):
  return AddDistribute(N, n, a, P_a) * LossDistribute(n,l,P_l)
  #return combination(n,l) * (P_l)**l*(1-P_l)**(n-l) * combination(N-n,a) * (P_a)**a*(1-P_a)**(N-n-a) 

def P_Success0(N,n,l,a,P_l,P_a):
  return Perr(N,n,0,0, P_l, P_a)

def P_Success1(N,n,l,a,P_l,P_a):
  P0 = P_Success0(N,n,l,a,P_l,P_a)
  P1 = Perr(N,n,0,1, P_l, P_a) + Perr(N,n,1,0, P_l, P_a)
  return P0 + P1

def P_Success2(N,n,l,a,P_l,P_a):
  P1 = P_Success1(N,n,l,a,P_l,P_a)
  P2 = Perr(N,n,1,1, P_l, P_a) + Perr(N,n,2,0, P_l, P_a) + Perr(N,n,0,2, P_l, P_a)
  return P1 + P2

def P_Success3(N,n,l,a,P_l,P_a):
  P2 = P_Success2(N,n,l,a,P_l,P_a)
  P3 = Perr(N,n,1,2, P_l, P_a) + Perr(N,n,2,1, P_l, P_a) + Perr(N,n,3,0, P_l, P_a) + Perr(N,n,0,3, P_l, P_a)
  return P_Success2(N,n,l,a,P_l,P_a) + P3
```

# Binomial Distribution of Error wihout Adjacent Elements included Cyclic Consideration 

Since the suggested shceme cannot correct adjacent erorrs, we need to edit the overall combination of error being detected in the distribution. The adjusted binomial distribution only applies to more than 1 error for having adjacent errors.
$$Prob(l, a, n, N, P_l, P_a) = \{ [\binom{n}{l} - n]  \times P_l^l(1-P_l)^{n-l} \} \times \{ [\binom{N-n}{a} - (N-n)] \times P_a^a (1-P_a)^{N-n-a} \} $$

```python
def LossDistributeWOAdjacent(n,l,P_l):
  return (combination(n,l) - n) * (P_l)**l*(1-P_l)**(n-l)

def AddDistributeWOAdjacent(N, n, a, P_a):
  return (combination(N-n,a) - (N-n)) * (P_a)**a*(1-P_a)**(N-n-a)

def PerrWOAdjacent(N,n,l,a, P_l, P_a):
  return AddDistributeWOAdjacent(N, n, a, P_a) * LossDistributeWOAdjacent(n,l,P_l)

# For loss or add error is 0 or 1
# n losses and 0 or 1 add, there is no adjacent errors
def PerrWOAdjacent_n0or1(N,n,l,a, P_l, P_a):
  return LossDistributeWOAdjacent(n,l,P_l) * AddDistribute(N, n, a, P_a)

# 0 or 1 loss and n add
def PerrWOAdjacent_0or1n(N,n,l,a, P_l, P_a):
  return LossDistribute(n,l,P_l) * AddDistributeWOAdjacent(N, n, a, P_a)

# 1 loss and 1 add, there is no adjacent errors
def PerrWOAdjacent_11(N,n,l,a, P_l, P_a):
  return  Perr(N,n,l,a, P_l, P_a)

#same as above definition
def PSuccessWOAdajacent0(N,n,l,a,P_l,P_a):
    return P_Success0(N,n,l,a,P_l,P_a)

def PSuccessWOAdajacent1(N,n,l,a,P_l,P_a):
  return P_Success0(N,n,l,a,P_l,P_a)  + Perr(N,n,0,1, P_l, P_a) + Perr(N,n,1,0, P_l, P_a)

#introduce adjacement error
def PSuccessWOAdajacent2(N,n,l,a,P_l,P_a):
  return PSuccessWOAdajacent1(N,n,l,a,P_l,P_a) + PerrWOAdjacent_11(N,n,1,1, P_l, P_a) + PerrWOAdjacent_n0or1(N,n,2,0, P_l, P_a) + PerrWOAdjacent_0or1n(N,n,0,2, P_l, P_a) # there is a problem for loss error to be 0 or 1 
def PSuccessWOAdajacent3(N,n,l,a,P_l,P_a):
  return PSuccessWOAdajacent2(N,n,l,a,P_l,P_a) + PerrWOAdjacent_0or1n(N,n,1,2, P_l, P_a) + PerrWOAdjacent_n0or1(N,n,2,1, P_l, P_a) + PerrWOAdjacent_n0or1(N,n,3,0, P_l, P_a) + PerrWOAdjacent_0or1n(N,n,0,3, P_l, P_a)
```

```python
# make a general expression for Binomial Distribution of Add Error
# (math.factorial(N-n) / (math.factorial(N-n-a)*math.factorial(a)))*(P_a)**a*(1-P_a)**(N-n-a)
# N, n, a, P_a
N, n, a, P_a = sy.symbols('N, n, a, P_a')
add = binomial(N-n,a)*(P_a)**a*(1-P_a)**(N-n-a)
```
