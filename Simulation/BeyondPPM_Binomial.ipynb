# Combination

$$\binom{n}{r} = \frac{n!}{r!(n-r)!}$$

Suppose we have a list [1,2,4], 3 photons with length of time bins 7.
The ways we can organize them is $3!$, and a selection of these 3 photons from a collection of this 7 time bins in which the order of selection does not matter is $\binom{n}{r}$. \\

https://www.math.net/combination



```python
def combination(n,r):
  return math.factorial(n) / (math.factorial(n-r)*math.factorial(r))

combination(3,2), combination(7,3), combination(14,4)
```

## Binomial Distribution of Loss Error
There are n photons in a time bin. The total probability of lossing $l$ photons and not losing $n-l$ photons

$$P(n, l, P_l) = \binom{n}{l} P_l^l(1-P_l)^{n-l} $$

$n$ is the number of photon. \
$l$ is the number of loss photon. \
$P_l$ is the probability of loss error. 

```python
def LossDistribute(n,l,P_l):
  return combination(n,l)*(P_l)**l*(1-P_l)**(n-l)
```

# Binomial Distribution of Error Probability

$$\text{ Binomail Distribution of Error} = \text{Binomial Distribution of Loss Error} \times \text{Binomial Distribution of Added Error} $$



$$Prob(l, a, n, N, P_l, P_a) = \binom{n}{l} P_l^l(1-P_l)^{n-l} \times \binom{N-n}{a} P_a^a (1-P_a)^{N-n-a}$$

$n$ = number of photon \
$l$ = number of photon lost \
$a$ = number of photon added \
$N$ = length of time bin \
$P_l$ = probability of lost error \
$P_a$ = probability of added error 

```python
def LossDistribute(n,l,P_l):
  return combination(n,l)*(P_l)**l*(1-P_l)**(n-l)  

def AddDistribute(N, n, a, P_a):
  return combination(N-n,a)*(P_a)**a*(1-P_a)**(N-n-a)

def Perr(N,n,l,a, P_l, P_a):
  return AddDistribute(N, n, a, P_a) * LossDistribute(n,l,P_l)
  #return combination(n,l) * (P_l)**l*(1-P_l)**(n-l) * combination(N-n,a) * (P_a)**a*(1-P_a)**(N-n-a) 

def P_Success0(N,n,l,a,P_l,P_a):
  return Perr(N,n,0,0, P_l, P_a)

def P_Success1(N,n,l,a,P_l,P_a):
  P0 = P_Success0(N,n,l,a,P_l,P_a)
  P1 = Perr(N,n,0,1, P_l, P_a) + Perr(N,n,1,0, P_l, P_a)
  return P0 + P1

def P_Success2(N,n,l,a,P_l,P_a):
  P1 = P_Success1(N,n,l,a,P_l,P_a)
  P2 = Perr(N,n,1,1, P_l, P_a) + Perr(N,n,2,0, P_l, P_a) + Perr(N,n,0,2, P_l, P_a)
  return P1 + P2

def P_Success3(N,n,l,a,P_l,P_a):
  P2 = P_Success2(N,n,l,a,P_l,P_a)
  P3 = Perr(N,n,1,2, P_l, P_a) + Perr(N,n,2,1, P_l, P_a) + Perr(N,n,3,0, P_l, P_a) + Perr(N,n,0,3, P_l, P_a)
  return P_Success2(N,n,l,a,P_l,P_a) + P3
```

# Binomial Distribution of Error wihout Adjacent Elements included Cyclic Consideration 

Since the suggested shceme cannot correct adjacent erorrs, we need to edit the overall combination of error being detected in the distribution. The adjusted binomial distribution only applies to more than 1 error for having adjacent errors.
$$Prob(l, a, n, N, P_l, P_a) = \{ [\binom{n}{l} - n]  \times P_l^l(1-P_l)^{n-l} \} \times \{ [\binom{N-n}{a} - (N-n)] \times P_a^a (1-P_a)^{N-n-a} \} $$

```python
def LossDistributeWOAdjacent(n,l,P_l):
  return (combination(n,l) - n) * (P_l)**l*(1-P_l)**(n-l)

def AddDistributeWOAdjacent(N, n, a, P_a):
  return (combination(N-n,a) - (N-n)) * (P_a)**a*(1-P_a)**(N-n-a)

def PerrWOAdjacent(N,n,l,a, P_l, P_a):
  return AddDistributeWOAdjacent(N, n, a, P_a) * LossDistributeWOAdjacent(n,l,P_l)

# For loss or add error is 0 or 1
# n losses and 0 or 1 add, there is no adjacent errors
def PerrWOAdjacent_n0or1(N,n,l,a, P_l, P_a):
  return LossDistributeWOAdjacent(n,l,P_l) * AddDistribute(N, n, a, P_a)

# 0 or 1 loss and n add
def PerrWOAdjacent_0or1n(N,n,l,a, P_l, P_a):
  return LossDistribute(n,l,P_l) * AddDistributeWOAdjacent(N, n, a, P_a)

# 1 loss and 1 add, there is no adjacent errors
def PerrWOAdjacent_11(N,n,l,a, P_l, P_a):
  return  Perr(N,n,l,a, P_l, P_a)

#same as above definition
def PSuccessWOAdajacent0(N,n,l,a,P_l,P_a):
    return P_Success0(N,n,l,a,P_l,P_a)

def PSuccessWOAdajacent1(N,n,l,a,P_l,P_a):
  return P_Success0(N,n,l,a,P_l,P_a)  + Perr(N,n,0,1, P_l, P_a) + Perr(N,n,1,0, P_l, P_a)

#introduce adjacement error
def PSuccessWOAdajacent2(N,n,l,a,P_l,P_a):
  return PSuccessWOAdajacent1(N,n,l,a,P_l,P_a) + PerrWOAdjacent_11(N,n,1,1, P_l, P_a) + PerrWOAdjacent_n0or1(N,n,2,0, P_l, P_a) + PerrWOAdjacent_0or1n(N,n,0,2, P_l, P_a) # there is a problem for loss error to be 0 or 1 
def PSuccessWOAdajacent3(N,n,l,a,P_l,P_a):
  return PSuccessWOAdajacent2(N,n,l,a,P_l,P_a) + PerrWOAdjacent_0or1n(N,n,1,2, P_l, P_a) + PerrWOAdjacent_n0or1(N,n,2,1, P_l, P_a) + PerrWOAdjacent_n0or1(N,n,3,0, P_l, P_a) + PerrWOAdjacent_0or1n(N,n,0,3, P_l, P_a)
```

For correcting up to 2 errors of either loss or added photon errors, the probability of success is that we may correct the received super block, $$P( \leq 2 \text{non-adjacent errors of either loss or add type}) = P(0,0) + P(1,0) + P(0,1) $$ 
$$
= \{ [\binom{n}{2} - n]  \times P_l^2(1-P_l)^{n-2} \} \times \{ [\binom{N-n}{a} - (N-n)] \times P_a^a (1-P_a)^{N-n-a} \} + P_l(1-P_l)^{n-1} \times P_a (1-P_a)^{N-n-1} + (1-P_l)^{n-2} \times P_a^2 (1-P_a)^{N-n-2} $$

### Analysis of Error Correction Performance for the (n=3, N=7) Code

This plot provides a powerful comparison of the error correction capabilities for the specific `n=3, N=7` code. It contrasts two distinct error models:

1.  **The Standard Model (`PSuccess` lines):** This model represents a theoretical "perfect" decoder that can correct *any* combination of up to `k` errors. It serves as a performance baseline or an upper bound on what is possible.
2.  **The Non-Adjacent Model (`PSuccessWOAdajacent` lines):** This model represents my proposed, more realistic scheme, which is designed to correct errors that do not occur in adjacent time bins.

The key takeaway is that my `k=2` non-adjacent scheme (brown line) provides a substantial improvement over a simple `k=1` scheme but pays a measurable performance penalty compared to a perfect `k=2` decoder (blue line).

---

### Detailed Analysis of Each Series

#### The Standard Model: Theoretical Baselines

These lines show what would be possible if you could correct *all* error patterns up to a certain number.

*   **`PSuccess 0` (Black 'x'):** This is the probability of **zero errors** occurring. It represents the most fragile case and its success rate drops very quickly. At a channel error probability of P=0.2, the chance of a perfectly error-free transmission is already below 20%.

*   **`PSuccess 1` (Red dots):** This is the probability of successfully correcting up to **one error**. By allowing for single-error correction, the system becomes significantly more robust. It holds up reasonably well until P0.2 and degrades from there.

*   **`PSuccess 2` (Solid Blue line):** This is the probability of correcting up to **two errors** of *any* kind (adjacent or not). This serves as the theoretical maximum performance for any scheme that aims to correct two errors.

*   **`PSuccess 3` (Dashed Green line):** This is the probability of correcting up to **three errors** of *any* kind. It is the most robust curve on the plot, representing the highest error-correction capability shown.

#### The Non-Adjacent Model: My Proposed Scheme

These lines represent the performance of my actual, constrained error correction scheme.

*   **`PSuccessWOAdajacent2` (Solid Brown line):** This is the central result for my 2-error correcting scheme.
    *   **Comparison to `PSuccess 1`:** The brown line is significantly above the red dots, which proves that my scheme is **effective**. Being able to correct non-adjacent 2-error patterns provides a major performance boost over only being able to correct single errors.
    *   **Comparison to `PSuccess 2`:** The brown line is slightly **below** the solid blue line. This gap is critical: it represents the **performance penalty** of my scheme. The space between the brown and blue lines represents the probability of encountering a 2-error pattern that is adjacent and therefore uncorrectable by my specific decoder.

*   **`PSuccessWOAdajacent3` (Dashed Gray line):** This shows the performance if you extend my scheme to correct up to 3 non-adjacent errors.
    *   **Comparison to `PSuccessWOAdajacent2`:** The gray line is slightly above the brown line, showing that adding 3-error correction capability improves performance, as expected.
    *   **Comparison to `PSuccess 2` and `PSuccess 3`:** The gray line is substantially below the dashed green line (`PSuccess 3`), again showing the penalty for the non-adjacency constraint. Interestingly, the gray line tracks the solid blue line (`PSuccess 2`) very closely.

### Key Insights and Conclusions

1.  **The Scheme Works:** The most important conclusion is that moving from a 1-error correcting capability to my non-adjacent 2-error scheme (from red to brown) yields a significant performance improvement.

2.  **Quantifying the Trade-off:** The plot beautifully visualizes the engineering trade-off. The "simplicity" of my scheme (by ignoring adjacent errors) comes at a cost, which is the performance gap between the brown and blue lines. You have sacrificed some error-correction power for a potentially simpler decoder.

3.  **Diminishing Returns:** The improvement from `PSuccessWOAdajacent2` (brown) to `PSuccessWOAdajacent3` (gray) is much smaller than the improvement from `PSuccess 1` (red) to `PSuccessWOAdajacent2` (brown). This suggests it might not be worth the added complexity to implement the 3-error decoder for this specific `(n=3, N=7)` code.

4.  **Performance Equivalence:** A fascinating insight is that `PSuccessWOAdajacent3` (gray) and `PSuccess 2` (blue) are nearly identical. This means that for the `n=3, N=7` code, my custom scheme needs to be able to handle up to 3 non-adjacent errors to achieve the same performance as a standard, ideal scheme that handles any 2 errors.

```python
import math
from itertools import islice
from functools import lru_cache

class CodeSequenceManager:
    """
    Manages the Mian-Chowla sequence to define the error correction code parameters.
    It pre-calculates the photon positions and the required number of time bins (N)
    for a given number of photons (n).
    """
    def __init__(self, max_photons=100):
        self._photon_positions = self._generate_sequence(max_photons)
        self._N_values = [sum(self._photon_positions[:i]) for i in range(1, max_photons + 1)]

    def _generate_sequence(self, count):
        """Generates the Mian-Chowla sequence (starting from a_1=0)."""
        aset, sset, k = set(), set(), 0
        sequence = []
        while len(sequence) < count:
            k += 1
            while any(k + an in sset for an in aset):
                k += 1
            sequence.append(k)
            sset.update(k + an for an in aset)
            aset.add(k)
        return [v - 1 for v in sequence] # Map to 0-indexed positions

    def get_N(self, n_photons):
        """Returns the total number of time bins (N) for n photons."""
        if not 0 < n_photons <= len(self._N_values):
            raise ValueError(f"n_photons must be between 1 and {len(self._N_values)}")
        return self._N_values[n_photons - 1]

# --- Initialize the manager ---
# We can now easily get N for any n. For example:
code_manager = CodeSequenceManager()
print(f"For n=3 photons, N = {code_manager.get_N(3)} time bins.")
print(f"For n=4 photons, N = {code_manager.get_N(4)} time bins.")
print(f"For n=5 photons, N = {code_manager.get_N(5)} time bins.")
print(f"For n=5 photons, N = {code_manager.get_N(6)} time bins.")
print(f"For n=5 photons, N = {code_manager.get_N(7)} time bins.")
print(f"For n=5 photons, N = {code_manager.get_N(8)} time bins.")
```

```python
@lru_cache(maxsize=None)
def combination(n, r):
    """Computes 'n choose r' with caching to improve performance."""
    if r < 0 or r > n:
        return 0
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

def combination_non_adjacent(k, r):
    """
    Computes combinations excluding patterns with adjacent elements in a cycle.
    The formula C(k,r) - k is a simplification that primarily applies to r=2.
    We will apply this logic for r >= 2.
    """
    if r < 2:
        return combination(k, r)
    # For r>=2, we subtract the k "adjacent" patterns (e.g., (1,2), (2,3)...(k,1))
    return combination(k, r) - k

def p_err(n, N, l, a, P_l, P_a):
    """Calculates the probability of exactly l losses and a additions (standard model)."""
    p_loss = combination(n, l) * (P_l**l) * ((1 - P_l)**(n - l))
    p_add = combination(N - n, a) * (P_a**a) * ((1 - P_a)**(N - n - a))
    return p_loss * p_add

def p_err_non_adjacent(n, N, l, a, P_l, P_a):
    """
    Calculates probability, but for correctable non-adjacent error patterns.
    We adjust the combination count for loss and add events separately.
    """
    # Adjust combination count for loss errors if l >= 2
    loss_comb = combination_non_adjacent(n, l)
    p_loss = loss_comb * (P_l**l) * ((1 - P_l)**(n - l))

    # Adjust combination count for add errors if a >= 2
    add_comb = combination_non_adjacent(N - n, a)
    p_add = add_comb * (P_a**a) * ((1 - P_a)**(N - n - a))
    
    return p_loss * p_add
```

```python
def p_success_upto_k(k_max, n, N, P_l, P_a, non_adjacent_model=False):
    """
    Calculates the total probability of success for correcting up to k_max total errors.
    A total error count is l + a.
    
    Args:
        k_max (int): The maximum number of total errors (loss + add) to correct.
        non_adjacent_model (bool): If True, use the non-adjacent error calculation.
    """
    total_prob = 0
    
    # Determine which error function to use
    error_func = p_err_non_adjacent if non_adjacent_model else p_err

    # Sum probabilities for all (l, a) pairs where l + a <= k_max
    for k in range(k_max + 1):
        for l in range(k + 1):
            a = k - l
            # Ensure we don't have more losses than photons or more adds than empty bins
            if l <= n and a <= (N - n):
                total_prob += error_func(n, N, l, a, P_l, P_a)
                
    return total_prob

# Example usage for n=3, N=7, correcting up to 2 errors
P = 0.1 # Example channel error probability
p_success_std = p_success_upto_k(2, 3, 7, P, P, non_adjacent_model=False)
p_success_non_adj = p_success_upto_k(2, 3, 7, P, P, non_adjacent_model=True)

print(f"\nFor P={P}:")
print(f"Standard model P(Success | <=2 errors) = {p_success_std:.4f}")
print(f"Non-adjacent model P(Success | <=2 errors) = {p_success_non_adj:.4f}")
```

### Analysis of Code Performance Across Different Sizes (n=3 to n=12)

This plot compares the performance of a family of non-adjacent error correction codes, all with the same fixed error-correcting power (`2 errors`). Each line represents a different code, defined by its number of photons (`n`) and the corresponding number of time bins (`N`).

The central question answered by this plot is: **How does the size of the code affect its reliability in a noisy channel?**

The results show a clear and dramatic trend: for a fixed error-correction capability, larger codes are significantly more fragile and less resistant to noise.

---

### Key Observations and Insights

#### 1. Inverse Relationship Between Code Size and Robustness

The most striking feature is the inverse relationship between the code size (`n`) and its performance.

*   The **`n=3, N=7` code (dark blue)** is by far the most robust. It maintains a high probability of success even when the channel error rate (`P`) approaches 0.3 or 0.4.
*   As `n` increases, the performance curves shift sharply to the left. The **`n=7, N=46` code (purple)**, for example, fails almost completely once the error rate exceeds P=0.2.
*   The codes with the highest `n`, such as **`n=12, N=332` (bright yellow)**, are extremely fragile. Their probability of success plummets to zero with even a tiny channel error rate (P < 0.05).

**Why does this happen?** As `n` and `N` increase, the total number of ways that errors can occur grows explosively. For instance, the total number of possible 2-error patterns is `combination(N, 2)`. While the number of correctable patterns also grows, it is dwarfed by the massive increase in uncorrectable patterns. A larger `N` simply provides overwhelmingly more opportunities for an uncorrectable error to occur, causing the system to fail.

#### 2. The "Waterfall" Effect and Performance Thresholds

Each curve exhibits a "waterfall" characteristic: it stays near a success probability of 1.0 for low error rates and then drops off very steeply. This cliff-like drop-off is typical of error correction codes and marks the **performance threshold**.

*   The location of this waterfall is a critical measure of a code's quality. A good code's waterfall occurs at a high error probability (far to the right).
*   This plot clearly shows that as `n` increases, the waterfall threshold moves to a much lower error probability (far to the left). This visually confirms the degradation in performance for larger codes.

#### 3. The Trade-off Between Information Rate and Reliability

This plot perfectly illustrates a fundamental concept in information theory. While not a formal calculation of code rate, we can think of using a higher `n` as an attempt to encode more information.

*   **Low Rate, High Reliability:** The `n=3` code has a low "density" of photons but is highly reliable. You can trust the decoded message even in a relatively noisy environment.
*   **High Rate, Low Reliability:** The `n=12` code attempts to pack more photons into its structure but is completely unreliable. It is only viable in a virtually noise-free channel.

This demonstrates that to achieve high reliability, one must often accept a lower information rate, and vice-versa.

### Overall Conclusion

For a system with a **fixed error-correcting power** (in this case, the ability to correct `2` non-adjacent errors), there is no "one-size-fits-all" best code. The optimal choice is entirely dependent on the expected noise level of the channel.

*   If the channel is known to be very noisy (e.g., P > 0.1), only the smallest codes like `n=3` or `n=4` are practical.
*   If the channel is guaranteed to be extremely clean (e.g., P < 0.01), one could potentially use a larger code like `n=10` to achieve a higher data throughput.

This analysis reveals a key limitation of the proposed scheme: its error-correcting power does not scale effectively with the size of the code. To make the larger codes (`n > 7`) viable in any realistic channel, the decoder would need to be enhanced to correct a greater number of errors (e.g., `k=3, 4, 5...`), which would likely increase its complexity.
